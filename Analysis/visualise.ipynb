```python
def processData(realLabels, genLabelsName):
    processedRealLabels=""
    processedGenLabels=""
    index=[]
    selectedIndex=[]
    indexCount=0
    n=0
    firstLine = False

    with open(genLabelsName, 'r') as file:
        for line in file:
            #print("Line at index ", indexCount,"  ", line  )
            # Skip first line
            if(firstLine==False):
                firstLine=True
            elif 'S' not in line and 'K' not in line:
                # If it doesn't contain "S", add the line to the processed content
                cleanedLine = line.replace("\n", "").replace(" ", "")
                processedGenLabels += cleanedLine
                selectedIndex.append(indexCount)
                indexCount+=1
                n+=1
                
            else:
                #indexes of lines that contain S in REAL file
                #print("Contains N!")
                index.append(indexCount)
                indexCount+=1
    
    #print(selectedIndex)
    indexCount=0
    lineCount=0
    with open(realLabels, 'r') as file:
        for line in file:
            lineCount+=1
            if ((indexCount in index) or not(indexCount in selectedIndex)):
                indexCount+=1
            else:
                indexCount+=1
                cleanedLine = line.replace("\n", "").replace(" ", "")
                for char in cleanedLine:
                    processedRealLabels += char
                          
    #print(processedRealLabels)
    #print("Original Val of n,",n)
    return processedRealLabels, processedGenLabels,n
```

```python
def PRFA(angryOnlyRealLabels, angryOnlyGenLabels):
    # Initialize counters for True Positives (TP), False Positives (FP), True Negatives (TN), False Negatives (FN)
    TP = FP = TN = FN = 0
    '''
    { 0: no emotion, 1: anger, 2: disgust, 3: fear, 4: happiness, 5: sadness, 6: surprise}
    # Calculate TP, FP, TN, FN
    for true_label, predicted_label in zip(angryOnlyRealLabels, angryOnlyGenLabels):
        if (true_label == '1' and (predicted_label == '1' or predicted_label == "2")):
            TP += 1
        elif true_label == '0' and predicted_label == '1':
            FP += 1
        elif true_label == '0' and (predicted_label == '0' or predicted_label== "2"):
            TN += 1
        elif true_label == '1' and predicted_label == '0':
            FN += 1
    '''
    for true_label, predicted_label in zip(angryOnlyRealLabels, angryOnlyGenLabels):
        if (true_label == '1' or true_label == '2' or true_label == '3' or true_label == '5') and predicted_label == '1':
            TP += 1
        elif (true_label != '1' and true_label != '2' and true_label != '3' and true_label != '5') and predicted_label == '1':
            FP += 1
        elif (true_label != '1' and true_label != '2' and true_label != '3' and true_label != '5') and predicted_label == '0':
            TN += 1
        elif (true_label == '1' or true_label == '2' or true_label == '3' or true_label == '5') and predicted_label == '0':
            FN += 1
    
    # Calculate Precision, Recall, F1 Score, and Accuracy
    try:
        precision = TP / (TP + FP)
        recall = TP / (TP + FN)
        f1_score = 2 * (precision * recall) / (precision + recall)
        accuracy = (TP + TN) / (TP + TN + FP + FN)
    except Exception as e:
        precision =9999999
        recall = 9999999
        f1_score =9999999
        accuracy = 9999999
        print(str(e))
        
    dataset = {
    "Precision": precision,
    "Recall": recall,
    "F1": f1_score,
    "Accuracy": accuracy,
    "FP": FP,
    "FN": FN,
    "TP": TP,
    "TN": TN,
    "#FTMessages": FP + FN + TP + TN,

    }

    print(dataset)
```

```python
genFile = "output54.0/output-54.0-2-N.txt"
realFile = "angry40.txt"
realLab, genLab, n = processData(realFile, genFile)
PRFA(realLab, genLab)

```
